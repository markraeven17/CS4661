```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data from CSV using the Pandas library
df = pd.read_csv('./crop_yield.csv')
df.head()
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
encoder = OneHotEncoder(sparse_output=False)
one_hot_encoded = encoder.fit_transform(df[categorical_columns])
one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(categorical_columns))
df_encoded = pd.concat([df, one_hot_df], axis=1)
df_encoded = df_encoded.drop(categorical_columns, axis=1)
df_encoded.head()
```

```python
df_encoded['Fertilizer_Used'] = df['Fertilizer_Used'].map({True: 1.0 , False: 0.0})
df_encoded['Irrigation_Used'] = df['Irrigation_Used'].map({True: 1.0 , False: 0.0})
X = df_encoded.drop(columns=['Yield_tons_per_hectare'])
y = df_encoded[['Yield_tons_per_hectare']]
```

```python
crops = ["Crop_Barley", "Crop_Cotton", "Crop_Maize", "Crop_Rice", "Crop_Soybean", "Crop_Wheat"]
barley_df = df_encoded[df_encoded["Crop_Barley"] == 1.0]
barley_df = barley_df.drop(crops, axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
barley_X = barley_df[['Rainfall_mm', 
        'Temperature_Celsius', 
        'Fertilizer_Used', 
        'Irrigation_Used', 
        'Days_to_Harvest',
        'Region_East',
        'Region_West',
        'Region_North',
        'Region_South',
        'Soil_Type_Clay',
        'Soil_Type_Sandy',
        'Soil_Type_Loam',
        'Soil_Type_Silt',
        'Soil_Type_Peaty',
        'Soil_Type_Chalky',
        'Weather_Condition_Cloudy',
        'Weather_Condition_Rainy',
        'Weather_Condition_Sunny',
         ]]
barley_y = barley_df[["Yield_tons_per_hectare"]]
barley_X_train, barley_X_test, barley_y_train, barley_y_test = train_test_split(barley_X, barley_y, test_size=0.25, random_state=3)
barley_scaler = StandardScaler()
barley_X_train = barley_scaler.fit_transform(barley_X_train)
barley_X_test = barley_scaler.transform(barley_X_test)
barley_linreg = LinearRegression()
barley_linreg.fit(barley_X_train, barley_y_train)
print(barley_linreg.intercept_)
print(barley_linreg.coef_)
barley_predict = barley_linreg.predict(barley_X_test)
barley_mse = mean_squared_error(barley_y_test, barley_predict)
barley_mae = mean_absolute_error(barley_y_test, barley_predict)
barley_r2 = r2_score(barley_y_test, barley_predict)

print("Mean Squared Error:", barley_mse)
print("Mean Absolute Error:", barley_mae)
print("(Accuracy)R² Score:", round(barley_r2*100),'%')
```
